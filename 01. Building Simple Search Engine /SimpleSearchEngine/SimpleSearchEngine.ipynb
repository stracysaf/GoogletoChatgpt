```python
import re
import unidecode
from bs4 import BeautifulSoup
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk.tokenize import word_tokenize
from nltk import download
import contractions

download('punkt')
download('stopwords')
download('omw-1.4')
```

```python
def preprocess_text(text):
    # 1. Remove HTML
    text = BeautifulSoup(text, "html.parser").get_text()

    # 2. Remove contradictions - assuming this means handling contractions (e.g., "don't" -> "do not")
    text = contractions.fix(text)
    
    # 3. Remove accented characters
    text = unidecode.unidecode(text)
    
    # 4. Remove non-alphanumeric characters (keeping spaces)
    text = re.sub(r'[^a-zA-Z0-9\s]', '', text)
    
    # 5. Convert to lowercase
    text = text.lower()
    
    # 6. Remove stopwords
    stop_words = set(stopwords.words('english'))
    word_tokens = word_tokenize(text)
    filtered_text = [word for word in word_tokens if word not in stop_words]
    text = ' '.join(filtered_text)
    
    
    # 7. Stemming
    stemmer = PorterStemmer()
    stemmed_text = [stemmer.stem(word) for word in text]
    text = ''.join(stemmed_text)

    # 8. Lemmatization
    lemmatizer = WordNetLemmatizer()
    lemmatized_text = [lemmatizer.lemmatize(word) for word in text.split(' ')]
    
    # Optional: Combine stemming and lemmatization
    # It's generally not recommended to use both, but if needed, uncomment the following lines:
    # stemmer = PorterStemmer()
    # stemmed_and_lemmatized_text = [stemmer.stem(word) for word in lemmatized_text]
    # text = ' '.join(stemmed_and_lemmatized_text)
    
    # If only lemmatization is needed:
    text = ' '.join(lemmatized_text)
    
    return text
```

```python
import gzip
from lxml import etree
from collections import Counter
import re
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize
import nltk

# Download necessary NLTK data (do this once)
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('stopwords')

def analyze_text(text):
    # Lowercase
    text = text.lower()
    
    # Tokenize
    tokens = word_tokenize(text)
    
    # Remove punctuation and non-alphabetic characters
    tokens = [token for token in tokens if token.isalpha()]
    
    # Remove stopwords
    stop_words = set(stopwords.words('english'))
    tokens = [token for token in tokens if token not in stop_words]
    
    # Lemmatization
    lemmatizer = WordNetLemmatizer()
    tokens = [lemmatizer.lemmatize(token) for token in tokens]
    
    return Counter(tokens)

def simplified_load_documents(file_path):
    doc_id = 0
    with gzip.open(file_path, 'rb') as f:
        for _, element in etree.iterparse(f, events=('end',), tag='doc'):
            yield {
                'ID': doc_id,
                'title': element.findtext('./title'),
                'url': element.findtext('./url'),
                'abstract': element.findtext('./abstract')
            }
            doc_id += 1
            element.clear()
            # Break after loading 10000 documents for demonstration; remove or adjust for full processing
            if doc_id == 100000:
                break

def build_index_and_document_details(documents):
    index = {}
    document_details = {}  # New mapping of ID to document details
    for doc in documents:
        doc_id = doc['ID']
        document_details[doc_id] = doc  # Store the document details
        
        term_frequencies = analyze_text(doc['abstract'])
        for term in term_frequencies:
            if term not in index:
                index[term] = set()
            index[term].add(doc_id)
    return index, document_details


if __name__ == '__main__':
    file_path = '../Playground/enwiki-latest-abstract.xml.gz'
    
    # Generate documents and build the index and document details
    documents_generator = simplified_load_documents(file_path)
    index, document_details = build_index_and_document_details(documents_generator)
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

def search_engine(document_details, query): 

  
    query_tokens = preprocess_text(query)

    matching_documents = []

    for doc_id, document in document_details.items():
        lower_title_tokens = preprocess_text(document['title'])
        lower_abstract_tokens = preprocess_text(document['abstract'])

      
        if any(token in lower_title_tokens for token in query_tokens) or \
           any(token in lower_abstract_tokens for token in query_tokens):
            matching_documents.append(document)

    return matching_documents
```

```python
def search(query, index):
    query_preprocess = preprocess_text(query)
    results = set()
    for word in query_preprocess.split(' '):
        if word in index:
            if not results:
                results = index[word]
            else:
                results = results.union(index[word])
        else:
            set()
    if results:
        print(results)
        results = list(results)
        results = results[:20]
        print(results)
        for doc_id in results:
            doc = document_details[doc_id]
            print("Title:", doc['title'])
            print("URL:", doc['url'])
            print("Abstract:", doc['abstract'])
            print("\n")
    else:
        print("No result found")
```
